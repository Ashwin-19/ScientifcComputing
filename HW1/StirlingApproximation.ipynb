#### $\text{Q2: Errors in Stirling's Approximation}$

```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import numpy.linalg as npla
import scipy.linalg as spla
```

```python
#function for factorial
def factorial(n):
    '''
    args -> n: natural number
    returns arr where arr[i] = i!
    '''
    arr = [1]*(n+1)
    i = 1
    
    while i <= n:
        arr[i] = i*arr[i-1] 
        i += 1
    
    return arr

# function for Stirling's Approximation
def approximation(n):
    '''
    formula
    n! = sqrt(2*pi*n)*((n/e)^n)
    returns n! using the above formula
    '''
    appx = np.sqrt(2*np.pi*n)
    appx *= np.power((n/np.exp(1)),n)
    return appx  
```

```python
x = np.arange(1,11,1)
factorial_accurate = np.array(factorial(10)[1:])
stirling_approximation = np.array([approximation(i) for i in range(1,11,1)])
```

```python
plt.scatter(x,factorial_accurate,label='factorial')
plt.scatter(x,stirling_approximation,label='stirling\'s approximation')
plt.legend(loc='upper left')
plt.title('Estimating Factorials using Stirling\'s approximation')
plt.xlabel('n')
plt.ylabel('n!')
plt.show()
```

```python
# calculating absolute errors
absolute_errors = np.absolute(factorial_accurate - stirling_approximation)
plt.scatter(x,absolute_errors,label='absolute error')
plt.legend(loc='upper left')
plt.title('Absolute Error upto n = 10')
plt.xlabel('n')
plt.ylabel('n!')
plt.show()
```

```python
# calculating relative errors
relative_errors = np.divide(absolute_errors, factorial_accurate)
plt.scatter(x,relative_errors,label='relative error')
plt.legend(loc='upper left')
plt.title('Relative Error upto n = 10')
plt.xlabel('n')
plt.ylabel('n!')
plt.show()
```

#### $\text{Q3(b):}$
$\text{The absolute errors increase whereas the relative errors decrease as n increases.}$
$\text{This is because the output of both the approximation function and the factorial}$
$\text{function depends exponentially on n and hence, their differences also vary}$
$\text{exponentially with changes in n.} \therefore \text{the absolute errors also increase exponentially.}$
