#### $\text{Q1: Polynomial Evaluation}$

```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import numpy.linalg as npla
import scipy.linalg as spla
```

```python
# enter formula for polynomial 
p = lambda x: (x-2)**9

# function to estimate polynomial given the coefficients
def p_estimate(coefficients, x):
    
    estimate = 0
    
    exp = len(coefficients) - 1
    
    for c in coefficients:
        estimate += c*(x**exp)
        exp -= 1
    
    return estimate     
    
```

```python
x = np.arange(1.920,2.080,0.001)
```

```python
P_graph = [p(i) for i in x]
coefficients = [1,-18,144,-672,2016,-4032,5376,-4608,2304,-512]
P_estimate_graph = [p_estimate(coefficients, i) for i in x]
```

```python
plt.plot(x,P_graph,label='polynomial')
plt.plot(x,P_estimate_graph,label='expansion')
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

#### $\text{Q3(a): }$
$\text{We would choose the polynomial (non expansion) method since it depends}$
$\text{continuously on the input data. As we can see from the plot that the expansion}$
$\text{generates, there are significant perturbations in p(x) with respect to small}$
$\text{changes in x. Hence, the expansion of p(x) is insensitive to perturbations due}$
$\text{to the approximations made.}$
