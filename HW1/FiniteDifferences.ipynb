#### $\text{Q4: Finite Differences and } \epsilon_{M}$

```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import numpy.linalg as npla
import scipy.linalg as spla
```

```python
f = lambda theta: np.tan(theta)

derivative = lambda x, h: (f(x+h)-f(x))/h
```

```python
x = np.linspace(0,16,17)
h_values = [np.power(10,-i) for i in x]
df_x = np.array([(1/np.power(np.cos(1),2))]*len(x))
df_x_estimate = np.array([derivative(1,h) for h in h_values])
```

```python
# calculating absolute error for finite 
absolute_error = np.log10(np.absolute(df_x - df_x_estimate))
x_axis = np.log10(h_values)
```

```python
plt.plot(x_axis,absolute_error)
plt.xlabel('log(h)')
plt.ylabel('log(abs error)')
plt.title('Finite Difference Formula')
plt.show()
```

#### $\text{Q4(a):}$
$\text{We notice that the error in finite difference approximation is bounded by } \frac{2 \epsilon}{h}$
$\text{and the computational error achieves a minima at h = }2\sqrt{\frac{\epsilon_{mach}}{M}}\text{ where M is an upper}$
$\text{bound on }|f^{''}(x)|$

$f^{''}(x) = 2Sec^{2}(x)tan(x)$

$\text{Using the tailor expansion we achieve the upper bound of M = c}$

$\text{So, we achieve a minima for a reasonable approximation of h}\approx10^{-8}$

$\text{And, as we can see in the plot as well, the log of absolute errors are the lowest at}$
$h\approx10^{-8}\approx \sqrt{\epsilon_{mach}}$

$\text{In this case, the total error increases for larger values of h due to increasing truncation}$
$\text{error and increases for smaller errors of h due to increasing rounding error}$

$\text{(Textbook Pg. 9, Classics in Applied Mathematics, Michael T. Heath.)}$

```python
cdappx = lambda x,h: (f(x+h)-f(x-h))/(2*h)
df_x_cdappx = np.array([cdappx(1,h) for h in h_values])
```

```python
# calculating absolute error for centered difference approximation
absolute_error_cda = np.log10(abs(df_x - df_x_cdappx))
```

```python
plt.plot(x_axis,absolute_error_cda)
plt.xlabel('log(h)')
plt.ylabel('log(abs error)')
plt.title('Centered Difference Approximation')
plt.show()
```

#### $\text{Q4(b):}$
$\text{For the centered difference approximation, the computed error attains a minima at h} \approx 10^{-7}$
$\text{Using the taylor theorem, we get a truncation error of } O(h^{2})\text{ as compared to }O(h)\text{ in the}$
$\text{finite difference approximation. Hence, we see a reduction in the truncation error.}$
